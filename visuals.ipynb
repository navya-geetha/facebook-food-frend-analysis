# **Face Book Food Trend Analysis-Visuals**

**Import libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Load dataset**

```python
df=pd.read_excel('/content/foodtrends_facebook_cleaned.xlsx')
```

**Likes Distribution**

```python
plt.figure()
df['likes_count'].hist()
plt.title('Likes Distribution')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```


**Comments Distribution**

```python
plt.figure()
df['comments_count'].hist()
plt.title('Comments Distribution')
plt.xlabel('Comments')
plt.ylabel('Frequency')
plt.show()
```

**Shares Distribution**

```python
plt.figure()
df['shares_count'].hist()
plt.title('Shares Distribution')
plt.xlabel('Shares')
plt.ylabel('Frequency')
plt.show()
```

**Likes vs Engagement**

```python
plt.figure()
plt.scatter(df['likes_count'], df['total_engagement'])
plt.title('Likes vs Total Engagement')
plt.xlabel('Likes')
plt.ylabel('Total Engagement')
plt.show()
```

**Post Type Count**

```python
plt.figure()
df['post_type'].value_counts().plot(kind='bar')
plt.title('Post Type Count')
plt.xlabel('Post Type')
plt.ylabel('Count')
plt.show()
```

**Time of Day Analysis**

```python
plt.figure()
df['time_of_day'].value_counts().plot(kind='bar')
plt.title('Posts by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.show()
```

**Correlation Heatmap**

```python
numeric_cols = df.select_dtypes(include='number').columns
corr = df[numeric_cols].corr()

plt.figure()
plt.imshow(corr)
plt.title('Correlation Heatmap')
plt.colorbar()

plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=90)
plt.yticks(range(len(numeric_cols)), numeric_cols)

plt.show()
```
